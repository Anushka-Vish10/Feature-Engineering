Probability Ratio Encoding

1.Probability of Survived based on Cabin--- Categorical Feature

2. Probability of Not Survived---1-pr(Survived)

3. pr(Survived)/pr(Not Survived)

4. Dictonary to map cabin with probability

5. replace with the categorical feature

```python
import pandas as pd
```

```python
# Using columns
df=pd.read_csv('titanic.csv',usecols=['Cabin','Survived']) 
df.head()
```

```python
### Replacing missing value with by filling with Missing
df['Cabin'].fillna('Missing',inplace=True)
df.head()
```

```python
df['Cabin'].unique() 
# Checking the unique value
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]  # taking on the first character of each unique categories 
df.head()
```

```python
df.Cabin.unique()
```

```python
# Probabiloty ratio encoding Generally used with target value

# Finding probability
prob_df=df.groupby(['Cabin'])['Survived'].mean()   # says that cabin valued w.r.t survived column

prob_df.head()
```

```python
# Converting it into dataframe
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
# Getting the probability of died as well

prob_df['Died']=1-prob_df['Survived']
```

```python
prob_df.head()
```

```python
# Finding the probability ratio

prob_df['Probability_ratio']=prob_df['Survived']/prob_df['Died']
prob_df.head()
```

```python
# Now converting this ratio to a dictionary
probability_encoded=prob_df['Probability_ratio'].to_dict()
```

```python
probability_encoded
```

```python
# now mapping the data with our data frame

df['Cabin_encoded']=df['Cabin'].map(probability_encoded)
df.head()
```
